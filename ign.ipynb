```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cd "/content/drive/MyDrive/Colab"
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab/ign.csv')
data[['number', 'score_phrase', 'title', 'url', 'platform', 'score', 'genre', 'editors_choice', 'release_year', 'release_month', 'release_day']]
```

```python
# Check the top platforms for games
# bar chart use for ranking

df2 = data.groupby(['platform'])['number'].count()[lambda x: x >= 500]
df2.plot(kind='bar', title='Top platforms for games', ylabel='Count number',
         xlabel='Platform', figsize=(6, 5))
plt.savefig('bar_chart.png', bbox_inches='tight')
```

```python
# Check the top platforms for games
# bar chart use for ranking

plt.figure(figsize=(15, 10))
data['platform'].value_counts(normalize=True).plot(kind='pie')
plt.savefig('pie_chart.png', bbox_inches='tight')
```

```python
# Platforms with highest releases by year
# bar chart use for ranking

plt.figure(figsize=(25, 20))

df4 = pd.to_numeric(data['release_year'], errors='coerce')[lambda x: x>2000].astype(int)
year_min = df4.min()
year_max = df4.max()

cnt = 0
for year in range(year_min, year_max+1):
  cnt += 1
  plt.subplot((year_max - year_min + 1 + 4) / 4, 4, cnt)
  df5 = data[data['release_year'] == str(year)].groupby(['platform'])['number'].count()
  df5.plot(kind='bar', title='Platforms with highest releases by year {}'.format(year), ylabel='Count number', xlabel='Platform')

plt.tight_layout()
plt.savefig('list_bar_chart.png', bbox_inches='tight')
```

```python
# Check the score distribution
# Histogram chart for distribution

df3 = pd.to_numeric(data['score'], errors='coerce')
plt.hist(df3, edgecolor="red", bins=50)
plt.savefig('histogram_chart.png')
plt.show()
```
